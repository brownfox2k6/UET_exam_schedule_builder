```python
import pandas, sqlite3

def show_invidiual_schedule(student_id: str | int):
  with sqlite3.connect("database.db") as conn:
    records = [*conn.cursor().execute(f"""
      SELECT s.student_id, s.course_code, co.course_name, s.day, s.slot, s.room, s.supervisor1, s.supervisor2
      FROM schedule s
      JOIN courses co ON s.course_code = co.course_code
      WHERE student_id = {student_id}
      ORDER BY s.day, s.slot
    """)]
  return pandas.DataFrame(records, columns=["Mã SV", "Mã MH", "Tên MH", "Ngày thi", "Ca thi", "Phòng thi", "CBCT 1", "CBCT 2"])
```
