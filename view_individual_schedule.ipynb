```python
import pandas, sqlite3

def show_invidiual_schedule(student_id: str | int):
  with sqlite3.connect("database.db") as conn:
    records = [*conn.cursor().execute(f"""
      SELECT en.student_id, en.section_code, se.course_code, co.course_name, s.day, s.slot
      FROM enrolments en
      JOIN sections se ON en.section_code = se.section_code
      JOIN courses co ON se.course_code = co.course_code
      JOIN schedule s ON s.course_code = co.course_code
      WHERE en.student_id = {student_id}
      ORDER BY s.day, s.slot
    """)]
  return pandas.DataFrame(records, columns=["Mã SV", "Mã LMH", "Mã MH", "Tên MH", "Ngày thi", "Ca thi"])
```
